```python
from subprocess import CREATE_NO_WINDOW
import urllib.request, urllib.error
from tkinter import messagebox
from selenium import webdriver
from tkinter import filedialog
import tkinter.ttk as ttk
from pathlib import Path
import tkinter as tk
import cv2
import os

# URLのチェック処理
def checkURL(url):
    
    try:
        f = urllib.request.urlopen(url)
        f.close()
        
        return url
    
    except:
        messagebox.showerror('エラー', 'URLが正しくありません')
        capture.release()
        
# カメラ設定の処理
def checkCAM(CameraName):
    
    if CameraName == "内臓カメラ":
        CameraNo = 0
    else:
        CameraNo = 1

    capture = cv2.VideoCapture(CameraNo)
    
    return capture

# パスが空白かどうかのチェック処理
def checkCAS(CasPath,capture):
    try:
        cascade = cv2.CascadeClassifier(CasPath)
        return cascade
    except:
        messagebox.showerror('エラー', 'カスケードファイルのパスが指定されていません')
        capture.release()
        
# パスが空白ではない場合のチェック処理
def checkCAS2(cascade,gray,capture):
    try:
        face = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30))
        return face
    except:
        messagebox.showerror('エラー', 'カスケードファイルが見つかりません')
        cv2.destroyWindow("Frame")
        capture.release()
        
# 分類器のパス取得
def dirdialog_clicked():
    
    fTyp = [("", "*.xml")]
    current_dir = os.path.abspath(Path().resolve())
    dir_path = filedialog.askopenfilename(filetypes = fTyp, initialdir=current_dir)
    entry_ws.set(dir_path)
    
# ブラウザ立ち上げる処理の関数
def web(url):
    
    global driver
    driver = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver.get(url)
    driver.maximize_window()
    
    return driver
    
# 起動時の処理の関数
def start():
    
    # テキストボックスの値を取得
    Cas = dir_entry.get()
    CameraName = CameraBox.get()
    urlc = UrlBox.get()
    
    # チェック処理
    url = checkURL(urlc)
    capture = checkCAM(CameraName)
    Cascade = checkCAS(Cas, capture)
    
    # フラグ作成
    firstloop = True
    
    # ブラウザ起動
    
    
    while(capture.isOpened()):
        
        ret,frame = capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        face = checkCAS2(Cascade, gray, capture)
        cv2.namedWindow("Frame", cv2.WINDOW_NORMAL)
        cv2.imshow("Frame", frame)
        
        if firstloop == True:
            driveri = web(url)
            firstloop = False
            
        for i in face:
            driver.quit()
            firstloop = False
            cv2.destroyWindow("Frame")
            capture.release()
            break
            
        if cv2.waitKey(1) & 0xFF == ord('q'):
            cv2.destroyWindow("Frame")
            capture.release()
            break
                   
# Tkinter
# メインウィンドウの設定
root = tk.Tk()
root.title("見せられないよ！")

# ウィンドウの大きさ指定
root.geometry("400x250")
root.minsize(width = 400, height = 250)
root.maxsize(width = 400, height = 250)

# ラベルの設定
label = tk.Label(root, text = "I won't show you!")
label.place(relx = 0.5,
            y = 35,
            anchor = tk.CENTER)

label = tk.Label(root,text = "カスケードファイルパス")
label.place(relx = 0.1,y = 60)

label = tk.Label(root, text = "URL")
label.place(relx = 0.1, y = 100)

label = tk.Label(root, text = "カメラ設定")
label.place(relx = 0.1, y = 140)

# URL指定
UrlBox = tk.Entry(root)
UrlBox.insert(0, "https://www.google.com/?hl=ja")
UrlBox.place(relx = 0.5,
             relwidth = 0.8,
             y = 130,
             anchor = tk.CENTER)

# カメラ指定
list1 = ("内臓カメラ", "外部カメラ")
CameraBox = ttk.Combobox(root,
                         values = list1,
                         state = "readonly")
CameraBox.set("内臓カメラ")
CameraBox.insert(0, 0)
CameraBox.place(relx = 0.1,
                relwidth = 0.2,
                y = 160)

# 分類器パス
entry_ws = tk.StringVar()
dir_entry = ttk.Entry(root, textvariable = entry_ws, width=20)
dir_entry.place(relx = 0.45,
                relwidth = 0.65,
                y = 90,
                x = -10,
                anchor = tk.CENTER)
dir_button = ttk.Button(root,
                        text="参照",
                        command = dirdialog_clicked)
dir_button.place(relx = 0.2,
                 width = 50,
                 y = 90,
                 x = 255,
                 anchor = tk.CENTER)

# 起動ボタンの設定
button1 = tk.Button(root,
                    text = "起動",
                    command = start,
                    bg = "#F5F5F5",
                    fg = "black")
button1.place(relx = 0.5,
              relwidth = 0.8,
              y = 210,
              anchor = tk.CENTER)

root.mainloop()
```
